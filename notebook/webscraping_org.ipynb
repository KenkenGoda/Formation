```python
# モジュールをインポート
import numpy as np
import pandas as pd
import re
import requests
from bs4 import BeautifulSoup
import pyprind
```

```python
# Excelファイルの読み出し
excel_book = pd.ExcelFile('/Users/shohei/Desktop/Shohei/Fresnel/Formation/dataset.xlsx')
df_raw = excel_book.parse('all').dropna(how='all')
```

```python
# planet nameのリストを抽出
planet_name = [name.replace(' ', '_') for name in df_raw['# name'].values]
```

```python
# 'exoplanets.org'から該当する惑星を検出した際の観測数、RMS、参考論文及びそのURLを抽出
pbar = pyprind.ProgBar(len(planet_name))
#NOBS = []
RMS = []
#REF = []
#URL = []

for name in planet_name:
    
    url = 'http://exoplanets.org/detail/{}'.format(name)
    res = requests.get(url)
    res.raise_for_status()
    soup = BeautifulSoup(res.text, "html.parser")
    script = soup.script.getText()
    
#    match_nobs = re.search(r'"NOBS"(.+)"OM"', script)  
#    if match_nobs:
#        nobs = re.sub(r'\D', '', match_nobs[1])
#        if nobs!='':
#            NOBS.append(int(nobs))
#        else:
#            NOBS.append(np.nan)
#    else:
#        NOBS.append(np.nan)
        
    match_rms = re.search(r'"RMS"(.+)"RR"', script)
    if match_rms:
        rms = re.findall(r'-?(([0-9]+)?\.[0-9]+|[0-9]+\.|([0-9]+|([0-9]+)?\.[0-9]+|[0-9]+\.)[eE][-+][0-9]+)', match_rms[1])
        if rms!=[]:
            RMS.append(float(rms[0][0]))
        else:
            RMS.append(np.nan)
    else:
        RMS.append(np.nan)
        
#    match_ref = re.search(r'"ORBREF"(.+)"ORBURL"', script)
#    if match_ref:
#        ref = re.search(r'"(.+)"', match_ref[1])
#        if ref!=[]:
#            REF.append(ref[1])
#        else:
#            REF.append(np.nan)
#    else:
#        REF.append(np.nan)
        
#    match_url = re.search(r'"ORBURL"(.+)"OTHERNAME"', script)
#    if match_url:
#        url = re.search(r'"(.+)"', match_url[1])
#        if url!=[]:
#            URL.append(url[1])
#        else:
#            URL.append(np.nan)
#    else:
#        URL.append(np.nan)
        
    pbar.update()
```

```python
# 観測数、RMS、参考論文及びそのURLのDFを作成しcsvファイルで保存
df = pd.DataFrame({'planet_name': df_raw['# name'].values,
                   'NOBS': NOBS,
                   'RMS': RMS,
                   'REF': REF,
                   'URL': URL
                  }).set_index('planet_name')
df.to_csv('/Users/shohei/Desktop/Shohei/Fresnel/Formation/webscraping.csv')
```
